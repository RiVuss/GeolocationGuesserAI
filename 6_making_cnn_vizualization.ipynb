```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Adjusted to ensure circular nodes and more input nodes
layers = [10, 15, 15, 15, 10, 8]  # Increased input nodes

# Function to calculate the positions of the nodes for a given layer (unchanged)
def get_layer_positions(layer_nodes, layer_idx, total_layers, x_spacing=2, y_spacing=1):
    x = np.full(layer_nodes, layer_idx * x_spacing)
    y = np.linspace(-(layer_nodes - 1) / 2, (layer_nodes - 1) / 2, layer_nodes) * y_spacing
    return x, y

# Initialize the plot
plt.figure(figsize=(10, 6))
plt.axis('off')

# Plot each layer and connect the nodes
for i in range(len(layers)):
    x, y = get_layer_positions(layers[i], i, len(layers))
    # Use 'edgecolor' to ensure nodes appear as circles
    plt.scatter(x, y, color='white', s=300, zorder=3, edgecolors='black', linewidths=0.5)
    
    if i > 0:  # Draw lines connecting nodes to the previous layer
        prev_x, prev_y = get_layer_positions(layers[i - 1], i - 1, len(layers))
        for px, py in zip(prev_x, prev_y):
            for cx, cy in zip(x, y):
                plt.plot([px, cx], [py, cy], color='black', linewidth=0.5, zorder=1)

plt.show()
```
